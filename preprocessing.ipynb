```python
import docx

def readtxt(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)
```

```python
text=""
for _ in os.listdir():
    if _.endswith(("doc","docx")) :
        text+=readtxt(_)
```
