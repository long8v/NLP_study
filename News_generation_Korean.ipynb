```python
# base="./scraping/Nate"
corpus=""
for portal in ["Nate","Daum"]:
    for section in os.listdir(base+"/"+portal):
        for file in os.listdir(base+"/"+portal+"/"+section):
            with open(base+"/"+portal+"/"+section+"/"+str(file),"r",encoding="utf-8") as fp:
                text=fp.read()
            for sentence in sent_tokenize(text):
                corpus+=sentence
```

```python
kkm=Kkma()
```

```python
POS_tagged=[kkm.pos(token) for token in word_tokenize(corpus) ]
```
